# SVM Classifier with linear and rbf kernel

```python
from sklearn.datasets.samples_generator import make_blobs
from matplotlib import pyplot as plt
import numpy as np
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
```

```python
data=read_csv(r'D:\project\mobile_price_range_data.csv')
```

```python
x,y=make_blobs(n_samples=125, centers=2, cluster_std=0.60, random_state=0)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=20, random_state=0)
```

```python
plt.scatter(x_train[:, 0], x_train[:, 1],c = y_train, cmap='winter')
```

```python
svc=SVC(kernel='linear')
svc.fit(x_train,y_train)
```

```python
plt.scatter(x_train[:, 0], x_train[:, 1],c = y_train, cmap='winter')

ax=plt.gca()
xlim=ax.get_xlim()

ax.scatter(x_train[:, 0], x_train[:, 1],c = y_train, cmap='winter',marker='s')

w=svc.coef_[0]
a=-w[0]/w[1]
xx=np.linspace(xlim[0],xlim[1])
yy=a*xx-(svc.intercept_[0]/w[1])
plt.plot(xx,yy)
plt.show()
```

```python
y_pred=svc.predict(x_test)
```

```python
#Confusion matrix

confusion_matrix(y_test,y_pred)
```

```python
#classification report

from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#best accuracy.

from sklearn.metrics import accuracy_score
print("Accuracy: ",accuracy_score(y_test,y_pred))
```
